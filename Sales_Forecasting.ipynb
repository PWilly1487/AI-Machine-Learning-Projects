Predict product sales based on advertising spend.

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Features: [Advertising budget]

x = np.array([
    [200000],

    [300000],

    [255000],

    [215000],

])

# Target: [Sales revenue]

y = np.array([
    190000,

    255000,

    330000,

    400000

])
```

```python
model = LinearRegression()

model.fit(x, y)
```

```python
print("Coefficients:", model.coef_)

print("Intercept:", model.intercept_)
```

# Task
Predict sales for new advertising spend values using the trained linear regression model, then visualize the original data, the new advertising spend values, and the predicted sales values on a scatter plot, and finally summarize the prediction results and visualization.

## Predict Sales

### Subtask:
Add code to create new advertising spend values and use the trained `model` to predict product sales for these new values.


**Reasoning**:
First, define a NumPy array for new advertising spend values, ensuring it has the correct shape for the model. Then, use the trained model to predict sales for these new values.



```python
new_advertising_spend = np.array([
    [220000],
    [310000],
    [270000]
])

predicted_sales = model.predict(new_advertising_spend)

print("New Advertising Spend:\n", new_advertising_spend)
print("\nPredicted Sales:\n", predicted_sales)
```

## Visualize Data and Predictions

### Subtask:
Visualize the original data points (advertising spend vs. sales), the new advertising spend values, and their corresponding predicted sales on a single scatter plot. Add a legend to differentiate between original and predicted data.

**Reasoning**:
To visualize the original data points and the predicted sales, I need to import the `matplotlib.pyplot` library. This is a common practice for plotting in Python.



```python
import matplotlib.pyplot as plt
print("Matplotlib imported successfully.")
```

**Reasoning**:
Now that matplotlib is imported, I will create a scatter plot to visualize the original advertising spend vs. sales, and the new advertising spend vs. predicted sales. This will include labels, a title, and a legend for clarity.



```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Original Data (Advertising Spend vs. Sales)')
plt.scatter(new_advertising_spend, predicted_sales, color='red', marker='x', s=100, label='Predicted Sales for New Advertising Spend')
plt.title('Advertising Spend vs. Sales (Original and Predicted)')
plt.xlabel('Advertising Spend')
plt.ylabel('Sales Revenue')
plt.legend()
plt.grid(True)
plt.show()
```

## Summarize Prediction Results and Visualization

### Subtask:
Summarize the prediction results, interpreting the relationship between advertising spend and sales, and briefly describe what the visualization shows regarding the model's performance and predictions.

### Summary of Prediction Results and Visualization

The linear regression model determined a relationship between advertising spend and sales. The model's coefficients indicate that for every unit increase in advertising spend, there is a corresponding change in sales. In this particular case, the coefficient is negative (-0.0477), suggesting an inverse relationship: as advertising spend increases, predicted sales tend to decrease slightly. The intercept (305321.58) represents the estimated sales when advertising spend is zero.

For the new advertising spend values of \[$220,000], \[$310,000], and \[$270,000], the model predicted sales of approximately \[$294,823.65], \[$290,529.05], and \[$292,437.76] respectively.

The scatter plot visually confirms this inverse relationship. The original data points show a trend where higher advertising spend is associated with lower sales, which is counterintuitive but what the model has captured from the provided data. The predicted sales for the new advertising spend values also fall along this inverse trend line. The plot clearly distinguishes between the original training data and the new predictions, showing how the model extrapolates or interpolates based on the learned relationship. The red 'x' marks representing predicted sales for new advertising spend values align with the general downward trend observed in the blue data points, reinforcing the model's interpretation of the data.
