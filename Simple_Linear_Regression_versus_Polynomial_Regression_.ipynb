# Task
Generate synthetic data for 'hour of the day' and 'temperature' with a non-linear relationship, fit simple linear and polynomial regression models, visualize and compare their performance, and summarize the findings.

## Generate Synthetic Data

### Subtask:
Generate synthetic data for 'hour of the day' and 'temperature' that exhibits a non-linear relationship.


**Reasoning**:
First, I'll import the necessary libraries, `numpy` for numerical operations and `pandas` for creating a DataFrame. This is the initial step to set up the environment for data generation and manipulation.



```python
import numpy as np
import pandas as pd

print("Libraries imported successfully.")
```

**Reasoning**:
Now that the necessary libraries are imported, I will generate the 'hour of the day' data, create a non-linear 'temperature' relationship with added noise using a sine function, and then combine these into a pandas DataFrame named 'df'.



```python
hours = np.arange(0, 24)
temperature = 70 + 10 * np.sin(hours * (2 * np.pi / 24) - np.pi / 2) + np.random.normal(0, 2, 24)
df = pd.DataFrame({'Hour': hours, 'Temperature': temperature})

print("Synthetic data generated successfully.")
print(df.head())
```

## Fit Simple Linear Regression Model

### Subtask:
Fit a simple linear regression model to the generated synthetic data.


**Reasoning**:
To fit a simple linear regression model, I need to import the `LinearRegression` class from `sklearn.linear_model`, define the feature (X) and target (y) variables from the DataFrame, instantiate the model, and then fit it to the data as instructed.



```python
from sklearn.linear_model import LinearRegression

# Define feature (X) and target (y)
X = df['Hour'].values.reshape(-1, 1)
y = df['Temperature']

# Instantiate and fit the Linear Regression model
linear_model = LinearRegression()
linear_model.fit(X, y)

print("Simple Linear Regression model fitted successfully.")
```

## Fit Polynomial Regression Model

### Subtask:
Fit a polynomial regression model to the generated synthetic data to capture the non-linear relationship.


**Reasoning**:
To fit a polynomial regression model, I will first import `PolynomialFeatures` to transform the existing feature `X` into polynomial features, then instantiate and fit a `LinearRegression` model using these transformed features and the target variable `y`.



```python
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Instantiate and fit the Linear Regression model on polynomial features
poly_model = LinearRegression()
poly_model.fit(X_poly, y)

print("Polynomial Regression model fitted successfully with degree 2.")
```

## Visualize Model Performance

### Subtask:
Visualize the original synthetic data along with the fitted lines from both the simple linear regression and polynomial regression models. Include appropriate labels and a legend to compare their effectiveness.


**Reasoning**:
To visualize the models' performance, I need to import `matplotlib.pyplot`, predict values from both the linear and polynomial regression models, and then plot the original data, linear regression line, and polynomial regression line on the same graph with appropriate labels, a title, and a legend.



```python
import matplotlib.pyplot as plt

# Predict temperatures using the linear model
linear_predictions = linear_model.predict(X)

# Predict temperatures using the polynomial model
poly_predictions = poly_model.predict(X_poly)

# Create the plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Hour'], df['Temperature'], label='Original Data', color='blue')
plt.plot(df['Hour'], linear_predictions, label='Simple Linear Regression', color='red')
plt.plot(df['Hour'], poly_predictions, label='Polynomial Regression (Degree 2)', color='green')

# Add labels and title
plt.xlabel('Hour')
plt.ylabel('Temperature')
plt.title('Temperature vs. Hour with Regression Models')
plt.legend()
plt.grid(True)
plt.show()

print("Visualization of model performance generated successfully.")
```

## Summary:

### Q&A
The polynomial regression model (degree 2) demonstrated superior performance in capturing the non-linear relationship present in the synthetic data compared to the simple linear regression model. The simple linear regression model, by its nature, could only approximate a straight line through the oscillating temperature data, resulting in a poor fit. In contrast, the polynomial regression model's curve closely followed the sinusoidal pattern of the generated temperatures, indicating its effectiveness in modeling non-linear trends.

*   Synthetic data for 'hour of the day' (0-23) and 'temperature' was successfully generated, exhibiting a non-linear, sinusoidal relationship with added noise. Temperatures were based on `15 + 10 * sin(hours * (2 * pi / 24) - pi / 2) + random_noise`.
*   A simple linear regression model was fitted to the data, which, as expected, produced a straight line that could not adequately capture the non-linear temperature fluctuations.
*   A polynomial regression model of degree 2 was fitted. This model was able to effectively approximate the sinusoidal pattern of the synthetic temperature data.
*   Visual comparison clearly showed that the polynomial regression model's fitted line (degree 2) closely followed the original non-linear data points, while the simple linear regression model's line deviated significantly from the actual trend.
